```python
# importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    """
    Compute the sigmoid of z

    Args:
        z (ndarray): A scalar, numpy array of any size.

    Returns:
        g (ndarray): sigmoid(z), with the same shape as z

    """

    g = 1 / (1 + np.exp(-z))

    return g
```

```python
x_train = np.array([0, 1, 2, 3, 4, 5]).reshape(-1, 1)
y_train = np.array([0, 0, 0, 1, 1, 1])
```

```python
print(x_train)
print(y_train)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)

w = lin_reg.coef_
b = lin_reg.intercept_
```

```python
print("w:", w)
print("b:", b)
```

```python
# Predict using linear regression and apply sigmoid
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(x_train, y_train)
```

```python
# Define boolean masks
pos = y_train == 1
neg = y_train == 0


# Scatter plot for positive (malignant) and negative (benign) classes
plt.scatter(
    x_train[pos],
    y_train[pos],
    marker="x",
    s=80,
    c="red",
    label="malignant",
    facecolors="none",
)
plt.scatter(
    x_train[neg],
    y_train[neg],
    marker="o",
    s=100,
    label="benign",
    facecolors="none",
    edgecolors="blue",
    lw=3,
)
# Defines a range of x values for plotting the sigmoid curve
x_range = np.linspace(x_train.min(), x_train.max(), 300).reshape(-1, 1)
y_range = log_reg.predict_proba(x_range)

# Plots sigmoid curve
plt.plot(x_range, y_range[:, 1],  color='green', label='Sigmoid Curve')



# Plots linear regression line
plt.plot(x_train, (w * x_train + b), color='orange', label='Linear Regression Output')

# Sets labels and legend
plt.ylabel("y")
plt.xlabel("Tumor Size")
plt.legend()

# Show plot
plt.show()
```
