### What Linear Algebra is?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

```python
t = np.linspace(0, 50, 1000) # start, finish, n points
```

Distance traveled by robber: d = 2.5t

```python
d_r = 2.5 * t
```

Distance travelled by sheriff: d = 3(t - 5) // Robber gets 5 min headstart

```python
d_s = 3 * (t - 5)
```

Solve the two equations for their intersect points

```python
def equations(vars):
    x, y = vars
    eq1 = (2.5 * x) - y
    eq2 = 3 * (x - 5) - y
    return [eq1, eq2]

# Initial guess for the solution
initial_guess = [0, 0]

# Solve the equations
x, y = fsolve(equations, initial_guess)
```

```python
fig, ax = plt.subplots()
plt.title("A bank robber caught")
plt.xlabel("time (in minutes)")
plt.ylabel("distance (in km)")
ax.set_xlim([0, 50])
ax.set_ylim([0, 100])
ax.plot(t, d_r, c="green")
ax.plot(t, d_s, c="brown")
plt.axvline(x=x, color="purple", linestyle="--")
plt.axhline(y=y, color="purple", linestyle="--")
plt.scatter(x, y, color="red")
plt.text(x, y, f'(time={x} mins, distance={y} km)', fontsize=10, verticalalignment='bottom', horizontalalignment='right')
```
